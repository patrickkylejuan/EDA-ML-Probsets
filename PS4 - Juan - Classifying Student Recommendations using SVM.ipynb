## 2. Build a classifier using an SVM to identify whether a comment results to a recommendation or not


```python
df_reviews = pd.read_sql('SELECT * FROM reviews', con=db_connection)
```

Student reviews for professors were generally positive i.e. most of the reviews in the portals consisted of recommendations.

```python
## Here we'll use the comments column as the input feature to the SVM
train_x, test_x, train_y, test_y = train_test_split(df_reviews["comments"], df_reviews["recommended"], test_size = 0.30, stratify = df_reviews["recommended"],random_state=17) #preserve the distribution
```

```python
## Using tfidf Vectorizer and vanilla SVM
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_train_x = vectorizer.fit_transform(train_x)
```

```python
from sklearn.svm import LinearSVC
classifier = LinearSVC()
classifier.fit(tfidf_train_x.toarray(), train_y)
```

```python
tfidf_test_x = vectorizer.transform(test_x) #applying tf-idf to the test data
print(tfidf_test_x.shape)
scores = cross_val_score(classifier, tfidf_test_x.toarray(), test_y, cv=5) #cross-validation
acc = scores.mean()
print("Accuracy: %0.2f percent" % (acc *100))
```

```python
# Checking out the confusion matrix from one random train test split 

predictions = classifier.predict(tfidf_test_x.toarray())

cf_matrix = confusion_matrix(test_y, predictions)
sns.heatmap(cf_matrix, annot=True, cmap='Blues', fmt='g')
pd.DataFrame(classification_report(test_y,predictions,output_dict=True))
```

### From the initial model above
we can see that our model performed generally well in terms of classifying positive reviews but did poorly on classifying non-recommendations. Let's see if we can try and improve the performance of the model in terms of the "Not recommended class" by adding the pros and cons columns of the db as features to the SVM classifier.

```python
### Trying out performance using a pros + cons + comments feature
combined_df = df_reviews.copy()
combined_df['total_comments'] = combined_df['pros'] + ' ' + combined_df['cons'] + ' ' + combined_df['comments']
combined_df['total_comments'] = combined_df['total_comments'].str.lower()
## Here we'll use the comments column as the input feature to the SVM
train_x, test_x, train_y, test_y = train_test_split(combined_df["total_comments"], combined_df["recommended"], test_size = 0.30, stratify = combined_df["recommended"],random_state=17)
vectorizer = TfidfVectorizer(stop_words='english') #dont necessarily have to clean data before fitting in tf idf -- punctuation inclusion depends on use case -- you can also add your custom stop words (to deal with other languages, read up more on common NLP practices)
tfidf_train_x = vectorizer.fit_transform(train_x)
classifier = LinearSVC()
classifier.fit(tfidf_train_x.toarray(), train_y)
tfidf_test_x = vectorizer.transform(test_x) #applying tf-idf to the test data
print(tfidf_test_x.shape)
scores = cross_val_score(classifier, tfidf_test_x.toarray(), test_y, cv=5) #cross-validation
acc = scores.mean()
print("Accuracy: %0.2f percent" % (acc *100))
# Checking out the confusion matrix from one random train test split 

predictions = classifier.predict(tfidf_test_x.toarray())

cf_matrix = confusion_matrix(test_y, predictions)
sns.heatmap(cf_matrix, annot=True, cmap='Blues', fmt='g')
pd.DataFrame(classification_report(test_y,predictions,output_dict=True))
```

### From the performance metrics of the (3-feature) model above,
we can see that the F1-scores improved by ~20% for the 0 class and the 1 class improved by ~ 1% too. So I guess we can say that the model performed marginally better with all the three columns used as features. Next steps for this would be hyperparameter optimization

## 4. Identify the top words for the pros and cons columns - Top words in terms of frequency and TF-IDF SCORE


```python
dummy = df_reviews.copy()
```

```python
english_stop_words = set(stopwords.words("english")) 
```

```python
def message_cleaning(message):
    Test_punc_removed = [char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in english_stop_words]
    final = ' '.join(Test_punc_removed_join_clean).lower()
    return final

dummy['cleaned_pros'] = dummy['pros'].apply(message_cleaning)
dummy['cleaned_cons'] = dummy['cons'].apply(message_cleaning)
```

```python
dummy.head()
```

```python
txt = dummy['cleaned_pros'].str.cat(sep = ' ');
```

```python
def output_df_frequent_words(df,column,top_n):
    df = df;
    txt = df[column].str.cat(sep = ' ');
    words = nltk.tokenize.word_tokenize(txt)
    word_dist = nltk.FreqDist(words)
    rslt = pd.DataFrame(word_dist.most_common(top_n), columns=['Word', 'Frequency'])
    return rslt

def make_wordcloud(df):
    df = df
    list_of_tuples = [tuple(x) for x in df.values]
    wordcloud = WordCloud().generate_from_frequencies(dict(list_of_tuples))
    plt.figure(figsize=(14,8))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()

def output_top_tfidf_score_words(df,column,top_n):
    txt = df[column].str.cat(sep = ' '); #convert the pandas df column (of sentences) into a bag of words
    tfidf_vec = TfidfVectorizer(stop_words="english")  #initialize vectorizer in the function using english stop words -- room for improvement: accomodate filipino stop words
    data = {'TEXT':[txt]}; # build a df from the bag of words to be input to the initialized tf idf vectorizer
    text_df = pd.DataFrame(data)
    transformed = tfidf_vec.fit_transform(text_df['TEXT'])  # fit the data 
    # get dictionary
    dictionary_tfidf = dict(zip(tfidf_vec.get_feature_names(), transformed.toarray()[0]))
    # sort dictionary
    sorted_transformed = dict(sorted(dictionary_tfidf.items(), key=lambda item: item[1], reverse = True))
    # build output df from dictionary
    data = [[word, tf_idf_score] for word, tf_idf_score in sorted_transformed.items()]
    result = pd.DataFrame(data, columns=['word', 'tf_idf_score'])
    result = result.head(top_n).copy()
    return result
    
```

## Top words in terms of Frequency and TF-IDF Score - PROS

```python
most_frequent_pros = output_df_frequent_words(dummy,'cleaned_pros',50)
most_frequent_pros.head(10) #top 10 words -- modify to show the top 10 words with the highest tf idf scores
```

```python
make_wordcloud(most_frequent_pros) # word cloud of 50 most frequent words
```

```python
most_important_pros = output_top_tfidf_score_words(dummy,'cleaned_pros',50)
most_important_pros.head(10) #top 10 words -- modify to show the top 10 words with the highest tf idf scores
# TF IDF pros and cons column, list it out
```
